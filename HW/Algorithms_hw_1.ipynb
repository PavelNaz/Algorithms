```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

№1

```python
n = X.shape[0]

eta = 1e-1 # ИЗМЕНЕНИЯ (новое)
n_iter = 260 # ИЗМЕНЕНИЯ (новое)

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

```python
n = X.shape[0]

eta = 1e-2 
n_iter = 1000 # ИЗМЕНЕНИЯ (новое)

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
#     for k in range(W.shape[0]):
#         W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))

    W -= eta * (1/n * 2 * np.dot(X.T, y_pred - y)) # ИЗМЕНЕНИЯ (новое)

    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,2)}')
```

```python
n = X.shape[0]

eta = 1e-2 
iter_num = 0 

min_weight_dist = 1e-8 

W = np.array([1, 0.5]) 

print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

weight_dist = np.inf 

while weight_dist > min_weight_dist: 
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    W_new = W - eta * (1/n * 2 * np.dot(X.T, y_pred - y)) 
    
    weight_dist = np.linalg.norm(W_new - W, ord=2)
    W = W_new.copy()
    
    if iter_num % 10 == 0:
        print(f'Iteration # ', iter_num, ': W_new = ', W, ', MSE = ', round(err,2), ', weight_dist = ', weight_dist)
    iter_num += 1
```
